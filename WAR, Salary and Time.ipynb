### Data Wrangling

```python
warBat1 = warBat.drop(columns=['playerID', 'team_ID', 'name_common', 'lg_ID', 'pitcher'])
```

```python
warBat1.info()
```

### Decision Tree Setup

```python
warBat1.dropna()
```

```python
warBat1.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
warBat1.fillna(warBat1.mean(), inplace=True)
```

```python
x = warBat1.drop('age', axis=1)
y = warBat1['age']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=76)
```

```python
decisionTree = DecisionTreeClassifier(random_state=76)
decisionTree.fit(x_train, y_train)
```

the decision tree classifier ran but I had to replace na with meean values not sure if this was a bad idea or not or how this might have affected the test data

```python
treePredictions = decisionTree.predict(x_test)
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test,treePredictions))
print(classification_report(y_test,treePredictions))
```
